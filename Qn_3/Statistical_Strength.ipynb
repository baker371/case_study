#### Qn. 3(a)

The company would like us to randomly survey customers who bought a Boom product. Select a random sample, with a sample size determined by a confidence level of 95% and a margin error of 5%. This sample should be stratified by region such that it is balanced for each. Export the list of randomly selected observations to a separate excel file.

**Selecting customers who bought boom products**

```python
df = data[data['Product'] == 'Boom']
df.shape
```

**Calculate the required sample size based on the confidence level and margin error:**

```python
confidence_level = 0.95
margin_error     = 0.05

# Calculate the total sample size
total_sample_size = int(np.ceil(1 / (margin_error**2)))

# Calculate the sample size per region
sample_size_per_region = int(np.ceil(total_sample_size / len(df['Region'].unique())))
```

**Stratified Sampling:**

```python
# Create an empty DataFrame to store the sampled data:
sampled_data = pd.DataFrame(columns=df.columns)
```

```python
# Iterate over each region
for region in df['Region'].unique():
  # Get all rows belonging to the current region
  region_data = df[df['Region'] == region]

  # Check if the region has enough data for sampling
  if len(region_data) > sample_size_per_region:
    # Randomly sample 'sample_size_per_region' rows from the region
    region_sample = region_data.sample(n=sample_size_per_region, random_state=42)

    # Append the region sample to the overall sampled data
    sampled_data = sampled_data.append(region_sample)
```

```python
# Specify the file path for the output Excel file
output_file_path = 'random_sampled_data.xlsx'

# Export the randomly selected observations to Excel
sampled_data.to_excel(output_file_path, index=False)
```

### Qn 3(b)

b.	The company would want to determine if the males have a higher repayment speed than the females. Using the sample data above, conduct a test statistic at an 88% confidence level and provide your conclusion.

**Performing a two-sample t-test**

```python
# Importing the statistics module
import scipy.stats as stats
```

```python
# Separate the data for males
male_data = sampled_data[sampled_data['Gender'] == 'MALE']
male_data.head()
```

```python
# Separate the data for males and drop missing or invalid values
female_data = sampled_data[sampled_data['Gender'] == 'FEMALE']
female_data.head()
```

```python
# Perform the two-sample t-test
statistic, p_value = stats.ttest_ind(male_data['Repayment Speed 2'], female_data['Repayment Speed 2'], equal_var=False)

# Set the desired confidence level
confidence_level = 0.88

# Check the p-value against the significance level
if p_value < (1 - confidence_level):
    conclusion = "Reject the null hypothesis. Males have a significantly higher repayment speed than females."
else:
    conclusion = "Fail to reject the null hypothesis. There is no significant difference in repayment speed between males and females."

# Print the results
print("Test Statistic:", statistic)
print("P-value:", p_value)
print("Conclusion:", conclusion)
```
